```python
telco_data = df.copy()
```

```python
telco_data['Total Charges'] = pd.to_numeric(telco_data['Total Charges'], errors='coerce')
telco_data.isnull().sum()
```

```python
telco_data.drop(columns=['Churn Category', 'Churn Reason'], inplace=True)
```

```python
telco_data.shape
```

```python
telco_data.tail()
```

```python
telco_data.isnull().sum()
```

```python
# Get the max tenure
print(telco_data['Tenure in Months'].max()) #72
```

```python
# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

telco_data['Tenure in Months_group'] = pd.cut(telco_data["Tenure in Months"], bins=range(1, 80, 12), right=False, labels=labels)
```

```python
telco_data['Tenure in Months_group'].value_counts()
```

```python
# Get the max age
print(telco_data['Age'].max()) #80
```

```python
# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 80, 12)]

telco_data['Age_group'] = pd.cut(telco_data["Age"], bins=range(1, 90, 12), right=False, labels=labels)
```

```python
telco_data.head()
```

```python
telco_data.drop(columns=['Customer ID', 'City','Latitude','Longitude','Zip Code','Tenure in Months','Age'], inplace=True)
```

```python
telco_data.head()
```

```python
for i, predictor in enumerate(telco_data.drop(columns=['Customer Status', 'Total Charges', 'Monthly Charge'])):
    plt.figure(i)
    sns.countplot(data=telco_data, x=predictor, hue='Customer Status')
```

```python
telco_data_dummies = pd.get_dummies(telco_data)
telco_data_dummies.head()
```

```python
sns.lmplot(data=telco_data_dummies, x='Monthly Charge', y='Total Charges', fit_reg=False)
```

```python
Mth = sns.kdeplot(telco_data_dummies['Monthly Charge'][telco_data_dummies["Customer Status"] == 0],
                  color="Red", shade=True)
Mth = sns.kdeplot(telco_data_dummies['Monthly Charge'][telco_data_dummies["Customer Status"] == 1],
                ax =Mth, color="Blue", shade= True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charge')
Mth.set_title('Monthly charges by churn')
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(telco_data_dummies.corr(), cmap="Paired")
```
